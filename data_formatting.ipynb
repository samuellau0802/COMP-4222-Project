Actually I dont know what is useful here. I will extract the birthplace of the people, as well as their degrees

```python
df_people = pd.read_csv(r"raw_data\people.csv")
df_people.head()
```

```python
df_relationship = pd.read_csv(r"raw_data\relationships.csv")
df_relationship.head()
```

```python
df_degree = pd.read_csv(r"raw_data\degrees.csv")
df_degree.head()
```

```python
df_people = df_people[["object_id", "birthplace"]]
df_relationship = df_relationship[["person_object_id", "relationship_object_id"]].rename(columns={'person_object_id': 'object_id'})
df_degree = df_degree[["object_id", "degree_type", "subject", "institution"]]
```

```python
df_people = df_people.merge(df_degree, on="object_id").groupby(['object_id']).agg({'degree_type': [list], 'birthplace': [set], 'degree_type': [list], 'subject': [list], 'institution': [list]}).reset_index()
df_people.columns = ["object_id", "degree_type", "birthplace", "subject", "institution"]
df_people['birthplace'] = df_people['birthplace'].apply(lambda x: list(x)[0])
df_people.head()
```
