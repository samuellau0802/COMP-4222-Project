## Choosing Useful Columns (+)

```python
from datetime import datetime
datetime.today().year
```

```python
import numpy as np
```

```python
df_companies["founded_year"] = df_companies["founded_at"].apply(lambda x: datetime.today().year-int(x[:4]) if x is not np.nan else np.nan)
df_companies = df_companies[["id", "normalized_name", "country_code", "funding_rounds", "funding_total_usd", "milestones", "relationships", "category_code", "founded_year"]]
df_companies
```

```python
df_companies = df_companies[["id", "normalized_name", "funding_rounds", "milestones", "relationships", "founded_year"]].join(pd.get_dummies(df_companies.country_code).join(pd.get_dummies(df_companies.category_code)))
df_companies.head()
```

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())
```

```python
for i in ["funding_rounds", "milestones", "relationships", "founded_year"]:
    df_companies[i] = normalize(df_companies[i])
df_companies.head()
```

```python
df_companies = df_companies.fillna(0)
```

```python
df_companies
```

```python
df_funds["founded_year"] = df_funds["fund_founded_at"].apply(lambda x: datetime.today().year-int(x[:4]) if x is not np.nan else np.nan)
df_funds = df_funds[["object_id", "fund_name", "country_code", "investment_rounds", "invested_companies", "fund_milestones", "fund_relationships", "founded_year"]]
df_funds
```

```python
df_funds = df_funds[["object_id", "fund_name", "investment_rounds", "invested_companies", "fund_milestones", "fund_relationships", "founded_year"]].join(pd.get_dummies(df_funds.country_code))
df_funds.head()
```

```python
for i in ["investment_rounds", "invested_companies", "fund_milestones", "fund_relationships", "founded_year"]:
    df_funds[i] = normalize(df_funds[i])
df_funds
```

```python
df_funds = df_funds.fillna(0)
```

```python
from sklearn import preprocessing
  
label_encoder_c = preprocessing.LabelEncoder()
label_encoder_c.fit(df_investment['funded_object_id'].values.tolist() + df_companies["id"].values.tolist())
df_investment['funded_object_id'] = label_encoder_c.transform(df_investment['funded_object_id'])
df_companies["id"] = label_encoder_c.transform(df_companies["id"])
```

```python
from sklearn import preprocessing
  
label_encoder_f = preprocessing.LabelEncoder()
label_encoder_f.fit(df_investment['investor_object_id'].values.tolist() + df_funds["object_id"].values.tolist())
df_investment['investor_object_id'] = label_encoder_f.transform(df_investment['investor_object_id'])
df_funds["object_id"] = label_encoder_f.transform(df_funds["object_id"])
```

```python
df_rounds = pd.read_csv("raw_data/funding_rounds.csv")
df_rounds
```

```python
df_rounds = df_rounds[["funding_round_id", "funding_round_code", "raised_amount", "participants", "is_first_round", "is_last_round"]]
df_rounds
```

```python
df_rounds = df_rounds.merge(df_investment[["funding_round_id", "funded_object_id", "investor_object_id"]], on="funding_round_id")
```

```python
df_rounds = df_rounds.join(pd.get_dummies(df_rounds.funding_round_code))
del df_rounds["funding_round_code"]
for i in ["raised_amount", "participants"]:
    df_rounds[i] = normalize(df_rounds[i])
```

```python
df_rounds = df_rounds[df_rounds["funded_object_id"].isin(df_companies["id"])]
df_companies = df_companies[df_companies["id"].isin(df_rounds["funded_object_id"])]

df_rounds = df_rounds[df_rounds["investor_object_id"].isin(df_funds["object_id"])]
df_funds = df_funds[df_funds["object_id"].isin(df_rounds["investor_object_id"])]
```

```python
startup_ind = df_companies.iloc[:,:1]
startup_ind = startup_ind.set_index('id')
startup_ind["new_index"] = list(range(len(startup_ind)))
startup_ind = startup_ind.to_dict()['new_index']

investor_ind = df_funds.iloc[:,:1]
investor_ind = investor_ind.set_index('object_id')
investor_ind["new_index"] = list(range(len(investor_ind)))
investor_ind = investor_ind.to_dict()['new_index']

df_companies['id'] = df_companies['id'].apply(lambda x: startup_ind[x])
df_funds['object_id'] = df_funds['object_id'].apply(lambda x: investor_ind[x])

df_rounds['funded_object_id'] = df_rounds['funded_object_id'].apply(lambda x: startup_ind[x])
df_rounds['investor_object_id'] = df_rounds['investor_object_id'].apply(lambda x: investor_ind[x])
```

```python
df_companies.to_csv("startups_formatted.csv", index=False)
df_funds.to_csv("investors_formatted.csv", index=False)
df_rounds.to_csv("funding_round_formatted.csv", index=False)
```
